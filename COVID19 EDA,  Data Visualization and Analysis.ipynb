```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range, upper_range

low, high=remove_outlier(df_country["TotalCases"])
df_country["TotalCases"]=np.where(df_country["TotalCases"]>high,high,df_country["TotalCases"])
df_country["TotalCases"]=np.where(df_country["TotalCases"]<low,low,df_country["TotalCases"])
```
